```python
data =pd.read_csv('householdtask3.csv')
```

```python
plt.scatter(data['year'],data['own'])
plt.title("scatter plot")
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

```python
#line chart
plt.plot(data['year'],data['own'])
plt.title("line chart")
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

```python
#bar chart
plt.bar(data['year'],data['own'])
plt.title("bar chart")
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

```python
#histogram
plt.hist(data['income'])
plt.title("histogram")
plt.show()
```
